# RMSD

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df_1 = pd.read_csv("1_PL_RMSD.csv")
df_2 = pd.read_csv("2_PL_RMSD.csv")
df_3 = pd.read_csv("3_PL_RMSD.csv")
df_4 = pd.read_csv("4_PL_RMSD.csv")

sns.set(style='darkgrid')
plt.figure(figsize=(6, 4))  # Tamaño de la figura

# Gráfica 1
sns.lineplot(data=df_1, x='Time(ns)', y='Prot_CA', label='CDK-2')

# Gráfica 2
sns.lineplot(data=df_2, x='Time(ns)', y='Prot_CA', label='CDK-6')

# Gráfica 3
sns.lineplot(data=df_3, x='Time(ns)', y='Prot_CA', label='Topo-1')

# Gráfica 4
sns.lineplot(data=df_4, x='Time(ns)', y='Prot_CA', label='VEGFR-2')

plt.title('RMSD de Cα')
plt.xlabel('Tiempo (ns)')  # Etiqueta del eje X
plt.ylabel('RMSD (Å)')  # Etiqueta del eje Y
plt.legend(loc="lower right") #,bbox_to_anchor=(1.2, 0.5))
dpi = 2000
plt.savefig('rmsd_MeDFO.svg', format='svg', dpi=dpi)

# Mostrar la gráfica
plt.show()
```
